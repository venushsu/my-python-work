# 1.用函數圖形模擬繪製麥當勞文藝logo

```python
noise = 0.1*np.random.randn(100)
```

```python
cl = np.random.randint(1,5,100)
```

```python
x1 = np.linspace(0,3.1,100)
y1 = np.sin(x1)
x2 = np.linspace(3.1,6.2,100)
y2 = np.sin(x1)
plt.scatter(x1,y1+noise,c=cl, s=50, alpha=0.6, cmap ="YlOrBr")
plt.scatter(x2,y2+noise,c=cl, s=50, alpha=0.6, cmap ="YlOrRd")
```

# 範圍標記

```python
from ipywidgets import interact 
```

```python
x =np.linspace(0,40,7000)
y = np.sin(x)*np.cos(x)
def move(n = 0):
    plt.scatter(x[y<0],y[y<0])
    plt.plot(x,y)
    
interact(move,n =(-40,40))
```
