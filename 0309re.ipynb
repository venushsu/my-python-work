```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.random.rand(10)
y =  np.random.rand(10)
```

```python
plt.scatter(x,y,s=80,c=x)
```

```python
def myplot(n=1):
    plt.scatter(x,y,c=x,s=100)
    t = np.linspace(-5,5,1000)
    plt.plot(t,0.5*np.sin(n*t)+0.3)
```

```python
myplot(1)
```
