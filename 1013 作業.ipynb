### 雙層選單 選歌曲類型 選歌 連結到youtube

```python
from IPython.display import YouTubeVideo
```

```python
from ipywidgets import interact
```

```python
twice= YouTubeVideo('3ymwOvzhwHs')
blackpink = YouTubeVideo('IHNzOHi8sJs')
jaychou = YouTubeVideo('HK7SPnGSxLM')
yogalin = YouTubeVideo('Mqr-kjvXsk8')
alanwalker = YouTubeVideo('YQRHrco73g4')
camila = YouTubeVideo('BQ0mxQXmLsk')
```

```python
def f(song):
    print(song)
```

```python
def g(kpop):
    print(kpop)
```

```python
def h(Mandarin):
    print(Mandarin)
```

```python
def q(English):
    print(English)
```

```python
interact(f,song = dict(kpop=(interact(g,kpop = dict(twice=(display(twice))
                      ,blackpink=(display(blackpink))))),Mandarin=(interact(h,Mandarin = dict(jaychou=(display(jaychou))
                      ,yogalin=(display(yogalin))))),English=(interact(q,English = dict(alanwalker=(display(alanwalker))
                      ,camila=(display(camila)))))) )
```
