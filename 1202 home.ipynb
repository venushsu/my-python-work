## contour 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(-20, 20)
y = np.linspace(0, 40)
[X,Y] = np.meshgrid(x, y)
Z = np.cos(X) + np.cos(Y)


plt.contourf(X,Y,Z, cmap="Set3", alpha=0.2)
plt.colorbar()
plt.contour(X,Y,Z, cmap="Pastel2")
```
